```python
import pandas as pd
import numpy as np
import matplotlib
import sklearn
```

```python
source_df = pd.read_csv(r"\Users\isaac\Code\GUH\Average-prices-Property-Type-2024-08.csv")
print(source_df)
```

```python
source_df.columns
```

```python
df = source_df[['Date',
                   'Region_Name'
                   ,'Area_Code'
                   ,'Detached_Average_Price'
                   ,'Semi_Detached_Average_Price'
                   ,'Terraced_Average_Price'
                   ,'Flat_Average_Price']]
```

```python
# Identify rows where Region_Name is 'City of London'
condition = df['Region_Name'] == 'City of London'

# Drop the identified rows
df = df.drop(df[condition].index)
```

```python
areaCode = df[['Area_Code', 'Region_Name']]
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Time'] = df['Date'].dt.year + (df['Date'].dt.month / 12)
df
```

```python
df = df.drop(columns=['Date', 'Area_Code'])
```

```python
df
df.drop(columns=["Year","Month"])
order = ['Time','Detached_Average_Price', 'Semi_Detached_Average_Price', 'Terraced_Average_Price', 'Flat_Average_Price', 'Region_Name']
df = df.reindex(columns=order)
df
```

```python
newColumns = []
for col in df['Region_Name'].unique():
    newColumn = df['Region_Name'] == col
    df[col] = newColumn.map({True: 1, False: 0})

df = df.drop(columns=['Region_Name'])
```
